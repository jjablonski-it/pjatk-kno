```python
import snscrape.modules.twitter as sntwitter
import pandas as pd
```

```python
tweet_topic = 'metaverse'
tweet_count = 100
```

```python
# Get tweets
tweets = []
for i, tweet in enumerate(sntwitter.TwitterSearchScraper(f'{tweet_topic} near:Gdansk within:10km').get_items()):
    if i >= tweet_count:
        break
    tweets.append(tweet)
len(tweets)
```

```python
tweets_df = pd.DataFrame(tweets, columns=['date', 'id', 'content', 'user', 'sourceLabel', 'lang', 'likeCount'])
tweets_df
```
